```python
# Se importa las librerias que se van a usar
import pandas as pd 
from xml.dom import minidom
from xml.etree.ElementTree import parse
```

```python
# Se guarda el xml en un document
documentUY = parse('xmlSentimiento/TASS2019_country_UY_train.xml')
print(documentUY)
```

```python
# Se saca en una lista los tweets y los sentimientos del archivo xml
listaUY[]
listaSentimiento=[]
for item in documentUY.iterfind('tweet'):
    listaUY.append(item.findtext('content'))
    listaSentimiento.append(item.find('sentiment').find("polarity").findtext('value'))
print(listaSentimiento)
```

```python
# Se presenta la lista de sentimientos
listaUY=[]
listaSentimiento=[]
for item in documentUY.iterfind('tweet'):
    listaUY.append(item.findtext('content'))
    listaSentimiento.append(item.find('sentiment').find("polarity").findtext('value'))
print(listaSentimiento)
```

```python
dataFrUy = pd.DataFrame({'Lista':listaUY})
dataFrUy
```

```python
# Se lo transforma el dataFrame a csv
dataFrUy = dataFrUy.to_csv("Content_UY.csv")
```
