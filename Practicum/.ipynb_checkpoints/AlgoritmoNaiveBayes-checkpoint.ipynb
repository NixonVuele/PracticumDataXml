```python
import pandas as pd
```

```python
country = pd.read_csv('Data/Countrys.csv')
```

```python
# Visualización de las 5 primeras filas
country.head()
```

```python
countryContSen = country[['content','sentiment']].copy()
```

```python
countryContSen
```

```python
# Conversion
countryContSen['sentiment'] = countryContSen.sentiment.map({'NONE':0,'N':1,'NEU':2, 'P':3})
```

```python
countryContSen.shape
```

```python
countryContSen
```

```python
# Importar el contador de vectorizacion e inicializarlo
from sklearn.feature_extraction.text import CountVectorizer
count_vector = CountVectorizer()
```

```python
# Visualizar del objeto'count_vector' que es una instancia de 'CountVectorizer()'
print(count_vector)
```

```python
count_vector.fit(countryContSen['content'])
names = count_vector.get_feature_names()
names
```

```python
doc_array = count_vector.transform(countryContSen['content']).toarray()
doc_array
```

```python
frequency_matrix = pd.DataFrame(data = doc_array, columns=names)
frequency_matrix
```

```python
# Dividir los datos en conjunto de entrenamiento y de test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test# Instantiate the CountVectorizer method
count_vector = CountVectorizer() = train_test_split(countryContSen['content'], countryContSen['sentiment'], random_state=1)
print('Number of rows in the total set: {}'.format(countryContSen.shape[0]))
print('Number of rows in the training set: {}'.format(X_train.shape[0]))
print('Number of rows in the test set: {}'.format(X_test.shape[0]))
```

```python
# Instantiate the CountVectorizer method
count_vector = CountVectorizer()
```

```python
# Fit the training data and then return the matrix
training_data = count_vector.fit_transform(X_train)
```

```python
# Transform testing data and return the matrix. Note we are not fitting the testing data into the CountVectorizer()
testing_data = count_vector.transform(X_test)
```

```python
from sklearn.naive_bayes import MultinomialNB
naive_bayes = MultinomialNB()
naive_bayes.fit(training_data, y_train)
```

```python
predictions = naive_bayes.predict(testing_data)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
print('Accuracy score: ', format(accuracy_score(y_test, predictions)))
```

```python
print('Precision score: ', format(precision_score(y_test, predictions, average='macro')))
```

```python
print('Recall score: ', format(recall_score(y_test, predictions,average='macro')))
```

```python
print('F1 score: ', format(f1_score(y_test, predictions,average='macro')))
```

```python
from sklearn.naive_bayes import MultinomialNB
naive_bayes = MultinomialNB()
naive_bayes.fit(training_data, y_train)
```

```python
print('Score:', format(f1_score(y_test, predictions,average='macro')))
```

```python
review = "Estoy muy Feliz"

review_vector = count_vector.transform([review]) # vectorizing
print(naive_bayes.predict(review_vector))
```

```python
import pickle

# pickling the vectorizer
pickle.dump(count_vector, open('Modelos/vectorizerBN.sav', 'wb'))

# pickling the model
pickle.dump(naive_bayes, open('Modelos/classifierNB.sav', 'wb'))
```
